```python
for num in range(3000) :
    if type(df_train.production_companies.iloc[num]) != float:
        var = str(df_train.production_companies.iloc[num][2:-2]).split(', ')
        for i in var:
            if 'name' in i:
                production_companies.append(i.split(': ')[-1])
```
